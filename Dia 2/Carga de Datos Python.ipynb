```python
def readCSV(fileName):
    file = open(fileName, "r")
    myFile = file.readlines()
    matrix = []
    for i in range(0,len(myFile)):
        matrix.append(myFile[i].rstrip("\n").split(","))
    return matrix
    
readCSV("avocado.csv")
```

### ¿Cuántos objetos hay? 

```python
len(readCSV("avocado.csv")) * len(readCSV("avocado.csv")[0])
```

### ¿Cuál es el valor de la variable 2 del objeto 15? 

```python
print(readCSV("avocado.csv")[15][2])
```

### ¿Cuál es el mínimo y máximo de la variable 3?

```python
var3Max = readCSV("avocado.csv")[1][2]
for i in range (2,len(readCSV("avocado.csv")[0])):
    if var3Max < readCSV("avocado.csv")[i][2]:
        var3Max = readCSV("avocado.csv")[i][2]
print("Max of Variable 3:")
print(var3Max)

var3Min = readCSV("avocado.csv")[1][2]
for i in range (2,len(readCSV("avocado.csv")[0])):
    if var3Min > readCSV("avocado.csv")[i][2]:
        var3Min = readCSV("avocado.csv")[i][2]

print("Min of Variable 3:")
print(var3Min)
```
